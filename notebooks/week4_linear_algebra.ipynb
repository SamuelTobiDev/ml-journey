```python
import numpy as np
from typing import List


weights = [-0.02, 0.005, -0.3, 1.2, -0.1]
rose    = [17, 71.28, 1, 1, 2]
jack    = [20, 7.25, 3, 0, 1]


# 1. Implement dot product from scratch
def dot_product(a: List[float], b: List[float]) -> float:
    return sum(a_i * b_i for a_i,b_i in zip(a,b))

print(f"Rose score (custom): {dot_product(weights, rose):.3f}")
print(f"Rose score (numpy): {np.dot(weights, rose):.3f}")
print(f"Jack score (custom): {dot_product(weights, jack):.3f}")
print(f"Jack score (numpy): {np.dot(weights, jack):.3f}")
```

```python
# 2. Implement cosine_similarity from scratch

king, queen = [0.9,0.1,0.8], [0.8,0.9,0.7]

def cosine_similarity(a: List[float], b: List[float]) -> float:
    magnitude_a = sum(x**2 for x in a) ** 0.5
    magnitude_b = sum(x**2 for x in b) ** 0.5
    return dot_product(a, b) / (magnitude_a * magnitude_b)

print(f"Cosine similarity (custom): {cosine_similarity(king, queen):.3f}")
print(f"Cosine similarity (numpy): {np.dot(king, queen) / (np.linalg.norm(king) * np.linalg.norm(queen)):.3f}")
```

```python
# 3. Implement matrix multiply from scratch

def matrix_multiply(A: List[List[float]], B: List[List[float]]) -> List[List[float]]:
    n,q,r = len(A[0]), len(B[0]) , len(A)

    if n != len(B):
        raise ValueError(f"Inner dimmension don't match: {n} vs {len(B)}")
    result = [[0 for _  in range(q)] for _ in range(r)]
    for i in range(r):
        for j in range(q):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

A = [[1,2,3],[4,5,6]]
B = [[7,8],[9,10],[11,12]]
print(f"Custom matrix mutltiply: {matrix_multiply(A, B)}")
print(f"NumPy matrix multiply: {(np.array(A) @ np.array(B)).tolist()}")
```
